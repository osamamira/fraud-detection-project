```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, RobustScaler,MinMaxScaler
```

```python
df=pd.read_csv('cleaned2.csv')
df
```

```python
df.drop(columns=['cc_num','trans_num','street'], inplace=True)
```

```python
df
```

```python
df['city'].unique()
```

```python
state_means = df.groupby('city')['is_fraud'].mean()
df['city'] = df['city'].map(state_means)
```

```python
df
```

# standardization step

## amt has outliers so, will scale with Robust scaler

```python
# Step 1: Initialize the RobustScaler
scaler = RobustScaler()

# Step 2: Reshape the 'amt' column for scaling (since RobustScaler expects a 2D array)
amt_values = df['amt'].values.reshape(-1, 1)

# Step 3: Fit the scaler on the 'amt' column and transform it
scaled_amt = scaler.fit_transform(amt_values)

# Step 4: Replace the original 'amt' column with the scaled values
df['amt'] = scaled_amt
```

## the columns need to scale are city and job only they do not have outliers so will use standard scaler

```python
# Step 1: Initialize the StandardScaler
scaler = StandardScaler()

# Step 2: Select the columns you want to scale
columns_to_scale = ['city', 'job']

# Step 3: Fit the scaler on the selected columns and transform them
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
```

```python
df
```

```python
# Assuming df is your DataFrame and 'age' is the column you want to scale

# Step 1: Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Step 2: Reshape the 'age' column for scaling (since MinMaxScaler expects a 2D array)
age_values = df['age'].values.reshape(-1, 1)

# Step 3: Fit the scaler on the 'age' column and transform it
scaled_age = scaler.fit_transform(age_values)

# Step 4: Replace the original 'age' column with the scaled values
df['age'] = scaled_age

# Now, df['age'] contains the scaled values
```
